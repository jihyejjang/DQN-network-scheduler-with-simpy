## graph

#### flow success rate

```python
#pexp
eps=[0.999]
for e in range(len(p)):
    if (e%200==0):
        eps.append(p[e])
```

```python
len(eps)
```

```python
len(success) #총 flow는 9431개 발생(왤케 적음..) -> learning rate 1e-3으로 했더니 17063개 발생
```

```python
rate=[]
mean=100
#flow_success=np.array(flow_success)
for i in range(int(len(success)/mean)):#200개당평균
    #print(success[mean*i:mean*(i+1)])
    rate.append(np.mean(success[mean*i:mean*(i+1)]))
    #print(np.mean(success[mean*i:mean*(i+1)]))
```

```python
x=range(int(len(success)/mean))
x2=range(int(len(eps)))
y=rate
y2=eps
plt.plot(x,y,x2,y2)
plt.xlabel('per 100 flows')
plt.ylabel('success rate')
plt.title('flow success rate')
plt.show()
```
